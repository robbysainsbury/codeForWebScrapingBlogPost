```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading in webscraped data
df = pd.read_csv("epidemics.csv")

#droppping columns I don't need
df = df.drop(df.columns[[0,6]], axis=1)

#displaying data
df
```

```python
#checking for nas in each column
df.isnull().sum()
```

```python
#checking for number of unique items in each column
df.nunique()
```

```python
df.dtypes
```

```python
#changing column types
df["Death toll (estimate)"] = pd.to_numeric(df["Death toll (estimate)"], errors='coerce')
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
```

```python

sns.scatterplot(data=df,x="Date",y="Death toll (estimate)", color='green')
plt.savefig('figure1.png')
```

```python
sns.distplot(df["Death toll (estimate)"],color='green')
plt.savefig('figure2.png')
```

```python
sns.displot(df["Disease"],color='green').set(xticklabels=[]).tick_params(bottom=False)
plt.savefig('figure3.png')
```

```python
sns.catplot(x='Disease', y = 'Death toll (estimate)', data = df, color='green').set(xticklabels=[]).tick_params(bottom=False)
plt.savefig('figure4.png')
```

```python
sns.displot(df["Location"],color='green').set(xticklabels=[]).tick_params(bottom=False)
plt.savefig('figure5.png')
```

```python
sns.catplot(x='Location', y = 'Death toll (estimate)', data = df, color='green').set(xticklabels=[]).tick_params(bottom=False)
plt.savefig('figure6.png')
```
